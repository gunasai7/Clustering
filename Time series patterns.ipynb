```

import pandas as pd
import numpy as np

from sklearn.cluster import KMeans
# Calculate silhouette_score
from sklearn.metrics import silhouette_score

import warnings
warnings.filterwarnings("ignore")

from google.colab import drive
drive.mount('/gdrive')
#Change current working directory to gdrive
%cd /gdrive
```

```
#Read files
trainfile = r'/gdrive/My Drive/CIS 508/Assignment 4/Sales.csv'
trainData = pd.read_csv(trainfile) #creates a dataframe
print(trainData.shape)
```

```

X_train = trainData.iloc[: ,55:107] #reading the normalized data
print(X_train.shape)
print(X_train.head())

```

```
#K-Means Clustering ========================================================================
#Default Mode - K=8
kmeans = KMeans()
kmeans.fit(X_train)
df_kmeans=X_train
kmeans_predict_train=kmeans.predict(df_kmeans)
df_kmeans['Cluster Number']=kmeans_predict_train
print(df_kmeans.shape)

print(silhouette_score(X_train, kmeans.labels_))
export_csv = df_kmeans.to_csv(r'/gdrive/My Drive/CIS 508/Assignment 4/DefaultKmeans.csv')
```

```
#K-Means Clustering ========================================================================
# K=20

kmeans = KMeans(n_clusters=20, init='random', algorithm='elkan')
kmeans.fit(X_train)

df_kmeans=X_train
kmeans_predict_train=kmeans.predict(X_train)

df_kmeans['Cluster Number']=kmeans_predict_train
print(silhouette_score(X_train, kmeans.labels_))
```

```
#K-Means Clustering ========================================================================
# K=2

kmeans = KMeans(n_clusters=2, init='random', algorithm='elkan')
kmeans.fit(X_train)

df_kmeans=X_train
kmeans_predict_train=kmeans.predict(X_train)

df_kmeans['Cluster Number']=kmeans_predict_train
print(silhouette_score(X_train, kmeans.labels_))
export_csv = df_kmeans.to_csv(r'/gdrive/My Drive/CIS 508/Assignment 4/HyperTuned Clusters-2.csv')
```
